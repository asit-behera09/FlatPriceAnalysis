```python
import numpy as np
import pandas as pd
```

## Data Cleaning

```python
file=pd.read_csv("Project_1.csv")
```

```python
file
```

```python
file["PriceInLakhs"]=file["PriceInLakhs"].apply(lambda x:x*100 if x<=10 else x*1)
file['PriceInLakhs']
```

```python
file["EstEMI"]=file["EstEMI"].apply(lambda x:x*100000 if x<=10 else x*1)

file["EstEMI"]
```

```python
file['PriceInLakhs']=file['PriceInLakhs'].astype('int')

file['EstEMI']=file['EstEMI'].astype('int')
```

```python
file.isna().sum()
```

```python
file["Facing"]=file["Facing"].fillna(file["Facing"].mode()[0])

file['Facing'].isna().sum()
```

```python
file["PropertyName"]=file["PropertyName"].fillna(file["PropertyName"].mode()[0])
```

```python
file
```

```python
file.isna().sum()
```

## Data Analysis

```python
# Adding a column of price per sqft 
file['PricePerSqft']=(file['PriceInLakhs']/file['Sqft'])*100000
```

```python
# Rounding the price per sqft and changing the data type from float to int
file['PricePerSqft']=file['PricePerSqft'].round().astype('int')
```

```python
# Stacked bar plot -two Categorical columns

# City Vs Price & With pricepr sqrft - which city has the highest price
#Locality vs price -which Locality has the highest price
# counts of Facing
#
```

```python
file
```

```python
file.info()
```

```python
file.describe()
```

```python
# Encoding the data column
file['BikeParking']=file['BikeParking'].apply(lambda x:0 if x==False else 1 )
file['CarParking']=file['CarParking'].apply(lambda x:0 if x==False else 1 )
```

```python
file.drop(columns='Unnamed: 0',inplace = True)
```

### Univariate analysis

```python
# Finding the average price of a flat in lakhs
file['PriceInLakhs'].mean()
```

```python
# Finding the average sqft of the flats 
file['Sqft'].mean()
```

```python
# Finding the average price per sqft 
file['PricePerSqft'].mean()
```

```python
# Finding the most frequent Facing of the flats
file['Facing'].mode()[0]
# By this we can say that most people prefer EAST facing flats.
```

### Multivariate analysis

```python
# Average sqft cost per locality 
file.groupby(['City','Locality'])[['Sqft','PricePerSqft']].median().astype('int')
```

```python
# What are the most available BHK for each locality
file.groupby(['City','Locality'])['BHK'].agg(lambda x: x.value_counts().idxmax())
```

```python
file.groupby(['City','Locality'])['Facing'].agg(lambda x: x.value_counts().idxmax())
```

```python
file.groupby(['City','Locality'])['BHK'].agg(lambda x: x.value_counts().idxmax())
```

```python
# Avg Sqft per each BHK
file.groupby(['Locality','BHK'])['Sqft'].median()
```

```python
# Comparing highest flat price to the average flat price of Locality as per BHK
file.groupby(['City','Locality','BHK'])['PriceInLakhs'].agg(['max','median'])
```

```python
# Which city has less flat price(in Lakhs) as per BHK
file.groupby(['BHK','City'])['PriceInLakhs'].min()
```

```python
grouped_data = file.groupby(['City', 'BHK']).size().reset_index(name='Count')

# Iterate over each city
for city in grouped_data['City'].unique():
    city_data = grouped_data[grouped_data['City'] == city]
    counts = city_data['Count']
    bhks = city_data['BHK']
    plt.figure(figsize=(4,3))
    plt.title(f'Apartment Distribution in {city}')
    plt.pie(counts, labels=bhks, autopct='%1.1f%%')
    plt.axis('equal')
    plt.legend()
    plt.show()
```
